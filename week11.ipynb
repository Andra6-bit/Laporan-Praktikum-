```python
import math
while True:
  print("1.menghitung luas lingkaran")
  print("2.menghitung faktorial angka")
  print("3.menghitung akar kuadrat")
  print("4.pembulatan ke atas")
  print("5.menghitung x pangkat y")
  pilihan = input("Masukkan pilihan anda(1/2/3/4/5): ")

  if pilihan == '1':
    r = float(input("Masukkan jari-jari lingkaran: "))
    luas = math.pi * r * r
    print("Luas lingkaran adalah: ", luas)
  elif pilihan == '2':
    p = int(input("Masukkan angka untuk do faktorial: "))
    faktorial = math.factorial (p)
    print("faktorial dari", p ,"adalah :" , faktorial)
  elif pilihan == '3':
    bilangan = float(input("Masukkan bilangan yang ingin dicari akar kuadratnya: "))
    akar = math.sqrt(bilangan)
    print("Akar kuadrat dari", bilangan, "adalah", akar)
  elif pilihan == "4":
    bilangan = float(input("Masukkan bilangan yang ingin dibulatkan ke atas: "))
    pembulatan = math.ceil(bilangan)
    print("Hasil pembulatan ke atas dari", bilangan, "adalah", pembulatan)
  elif pilihan == "5":
    x = float(input("Masukkan nilai x: "))
    y = float(input("Masukkan nilai y: "))
    hasil = math.pow(x, y)
    print(x, "pangkat", y, "adalah", hasil)
    break
  else:
    print("Pilihan tidak valid. Silahkan coba lagi.")



```
