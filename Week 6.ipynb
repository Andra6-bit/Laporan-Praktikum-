```python
print('Show Perfect number fom 1 to 100')
n = 2
while n <= 100:
    x_sum = 0
    for i in range(1, n):
        if n % i == 0:
            x_sum += i
    if x_sum == n:
        print('Perfect number:', n)
    n += 1
```

```python

from math import factorial

n = int(input("Masukkan Jumlah Baris: "))
for i in range(n):
    for j in range(n-i+1):

        print(end=" ")

    for j in range(i+1):

        print(factorial(i)//(factorial(j)*factorial(i-j)), end=" ")

    print()
```

```python
nilai = int(input("Masukkan Baris yang Diinginkan: "))

baris = [1]
jumlah_baris = 1

for i in range(1, nilai):
    baris_baru = [1]
    for j in range(1, len(baris)):
        nilai_baru = baris[j - 1] + baris[j]
        baris_baru.append(nilai_baru)
        jumlah_baris += nilai_baru
    baris_baru.append(1)
    baris = baris_baru

print("Angka Satu Baris Segitiga Pascal:")
for angka in baris:
      print(angka, end=" ")
```
