Buatlah sebuah program login sebuah sistem. User diminta untuk memasukkan username beserta password

1. User sukses login apabila username = admin, dan password = password, maka berhasil login sebagai admin
2. Apabila username = admin, dan password = 12345, maka muncul output berupa teks yang menyatakan user perlu merubah password karena kata sandi lemah
3. Apabila username = guest, dan password = guest, maka berhasil login sebagai guest
4. Apabila username = guest, dan password selain guest, maka muncul output berupa password salah
5. selain rules di atas, maka muncul output berupa user tidak ditemukan

```python
  print ('LOGIN')

  username = input("Masukkan Nama : ")
  password = input('Masukkan Password : ')

  if username == 'admin' and password == 'password':
    print('berhasil login sebagai admin')
  elif username == 'admin' and password == '12345':
    print('User perlu merubah password karena kata sandi lemah')
  elif username == 'guest' and password == 'guest':
    print('Berhasil Login Sebagai Guest')
  elif username == 'guest' and password != 'guest':
    print('Password Salah')
  else :
    print('User tidak ditemukan')


```

Buatlah sebuah program untuk mengecek apakah angka yang dimasukkan user merupakan bilangan genap atau ganjil, dengan ketentuan sebagai berikut:
(Quiz ini silahkan diupload di Elita maksimal di akhir waktu perkuliahan)

1. User perlu memasukkan sebuah angka
2. Program akan cek apakah angka yang dimasukkan merupakan bilangan bulat, apabila bilangan bulat maka program akan berlanjut pada proses selanjutnya. Namun apabila bukan bilangan bulat, maka program akan memberikan output bahwa angka yang dimasukkan bukan merupakan bilangan bulat, dan program selesai.
3. Apabila bilangan bulat, maka akan dicek apakah bilangan tersebut bilangan negatif, nol, atau bilangan positif. Apabila bilangan negatif maka program akan memberikan output bahwa angka yang dimasukkan merupakan bilangan negatif, dan program selesai. Apabila angka yang dimasukkan adalah nol, maka program akan memberikan output bahwa angka yang dimasukkan merupakan nol, dan program selesai.
4. Apabila bilangan yang dimasukkan adalah bilangan positif, maka akan dicek apakah bilangan yang dimasukkan adalah bilangan ganjil atau genap. Apabila bilangan ganjil, maka program akan memberikan output bahwa angka yang dimasukkan merupakan bilangan ganjil, dan program selesai. Apabila bilangan genap, maka program akan memberikan output berupa angka yang dimasukkan merupakan bilangan genap, dan program selesai.

```python
print(' Mengecek apakah bilangan genap atau ganjil')

while True:
  try:
    angka = int(input('Masukkan sebuah angka : '))
    break
  except ValueError:
    print('Bukan bilangan bulat')
    break


if angka > 0:
  if angka % 2 == 0:
    print('Bilangan Genap')
  else:
    print('Bilangan Ganjil')
elif angka == 0:
  print('Bilangan Nol')
else:
  print('Bilangan Negatif')




```

while True
Fungsi: Membuat loop tak terbatas yang akan terus berjalan sampai ada perintah break yang dieksekusi.
Kegunaan:
Meminta input dari pengguna berulang kali sampai input yang valid diberikan.
Melakukan suatu tindakan terus-menerus sampai kondisi tertentu terpenuhi.
Membuat server atau program yang selalu berjalan.

try
Fungsi: Membungkus satu blok kode yang berpotensi menimbulkan error.
Kegunaan:
Menangkap error yang mungkin terjadi dan memberikan penanganan yang sesuai.
Memastikan program tetap berjalan meskipun terjadi kesalahan.

except ValueError
Fungsi: Menangkap kesalahan khusus ValueError.
Kegunaan:
Menangani situasi di mana terjadi kesalahan saat mencoba mengkonversi suatu nilai ke tipe data yang tidak sesuai (misalnya, mengkonversi string yang bukan angka menjadi integer).

```python
print('Tahun Kabisat atau Bukan')

tahun = int(input('Msukkan tahun: '))

if tahun % 4 == 0:
  if tahun % 100 == 0:
    if tahun % 400 == 0:
      print('Tahun Kabisat')
    else:
      print('Bukan Tahun Kabisat')
  else:
    print('Bukan Tahun Kabisat')
else:
  print('Bukan Tahun Kabisat')
```
