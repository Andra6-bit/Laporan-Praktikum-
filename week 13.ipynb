```python
class angka:
  def __iter__(self):
    self.k = 1
    return self

  def __next__(self):
    if self.k <= 20:
      b = self.k
      self.k += 1
      return b
    else:
      raise StopIteration

kelas = angka()
for x in kelas:
  print(x)
```

```python
class AlatTulis:
    def __init__(self, nama, stok, harga_satuan):
        self.__nama = nama
        self.__stok = stok
        self.__harga_satuan = harga_satuan
        self.__harga = stok * harga_satuan

    def set_data(self, nama, stok, harga_satuan):
        self.__nama = nama
        self.__stok = stok
        self.__harga_satuan = harga_satuan
        self.__harga = stok * harga_satuan

    def get_data(self):
        return {
            "Nama": self.__nama,
            "Stok": self.__stok,
            "Harga Satuan": self.__harga_satuan,
            "Harga": self.__harga
        }

    def set_total_harga(self, jumlah_terjual):
        if jumlah_terjual > self.__stok:
            raise ValueError("Jumlah terjual melebihi stok!")
        self.__stok -= jumlah_terjual
        return jumlah_terjual * self.__harga_satuan

    def get_total_harga(self):
        return self.__stok * self.__harga_satuan

bolpoint = AlatTulis("Bolpoint", 10, 2000)
pensil = AlatTulis("Pensil", 10, 1000)
penghapus = AlatTulis("Penghapus", 10, 500)

print(bolpoint.get_data())
print(pensil.get_data())
print(penghapus.get_data())

print("\nPenjualan 3 bolpoint:")
total_bayar_bolpoint = bolpoint.set_total_harga(3)
print(f"Total Bayar Bolpoint: Rp. {total_bayar_bolpoint}")
print(bolpoint.get_data())
```
