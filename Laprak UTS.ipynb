```python
npm = input('Masukkan NPM anda : ')
genap = 0
ganjil = 0
for i in range(len(npm)):
    if int(npm[i]) % 2 == 0:
        genap += 1
    else:
        ganjil += 1
print(f'Jumlah digit genap : {genap}')
print(f'Jumlah digit ganjil : {ganjil}')
```

```python
baris = int(input('Masukkan jumlah baris : '))
for i in range(1,baris + 1):
    for j in range(1,i+1):
        print(i*j, end =' ')
    print()
```
