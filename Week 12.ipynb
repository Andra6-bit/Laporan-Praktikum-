```python
try:
  baris = input('Masukkan jumlah baris : ')
  b = int(baris)

  for i in range(1,b + 1):
      for j in range(1,i+1):
          print(i*j, end =' ')
      print()
except ValueError:
  if not baris.strip():
    print("tidak boleh kosong")
  else:
    print('Hanya boleh memasukkan angka')
```

```python

try:
  npm = input('Masukkan NPM anda : ')
  if not npm.isdigit():
        raise ValueError("Input harus berupa angka")

  genap = 0
  ganjil = 0
  for i in range(len(npm)):
    if int(npm[i]) % 2 == 0:
      genap += 1
    else:
      ganjil += 1
  print(f'Jumlah digit genap : {genap}')
  print(f'Jumlah digit ganjil : {ganjil}')
except ValueError as e:
  print(e)
```
