```python
panjang = 20.5
lebar = 25
tinggi = 10

keliling = 2*(panjang+lebar)
luas = panjang*lebar
volum = panjang*lebar*tinggi

print("panjang = " + str (panjang))
print("lebar = " + str (lebar))
print('tinggi = ' + str(tinggi))

print("keliling  = " + str(keliling))
print("luas = " + str(luas))
print("volume = " + str (volum))
```
